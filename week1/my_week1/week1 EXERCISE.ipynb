and responds with an explanation. This is a tool that you will be able to use yourself during the course!

```python
# imports
import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display
```

```python
# constants
load_dotenv(override=True)
GEMINI_API_KEY =os.getenv("GEMINI_API_KEY")
```

```python
# set up environment
gemini=OpenAI(base_url="https://generativelanguage.googleapis.com/v1beta/openai/", api_key=GEMINI_API_KEY)

question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""
messages=[
  {"role": "system" , "content": "You are a technical expert. Please answer the question in detail and respond only in Markdown formatted text."},
  {"role":"user", "content": question}
  
]
stream=gemini.chat.completions.create(
  model="gemini-2.5-flash-lite",
  messages=messages,
  stream=True
)
full_response=""
for chunk in stream:
  display(Markdown(chunk.choices[0].delta.content or ""))
  full_response+=chunk.choices[0].delta.content



```
